## 2.3 –°—Ç–æ–ª–±—á–∞—Ç—ã–µ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã –∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –¥–ª—è –≥—Ä—É–ø–ø—ã –ø–æ–ø–∞–≤—à–∏—Ö –≤ –æ—Ç—Ç–æ–∫ –∏ –¥–ª—è –≥—Ä—É–ø–ø—ã –æ—Å—Ç–∞–≤—à–∏—Ö—Å—è.

–ü–æ—Å—Ç—Ä–æ–∏–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –¥–ª—è —Ç–µ—Ö, –∫—Ç–æ —É—à—ë–ª (–æ—Ç—Ç–æ–∫) –∏ —Ç–µ—Ö, –∫—Ç–æ –æ—Å—Ç–∞–ª—Å—è (–Ω–µ –ø–æ–ø–∞–ª–∏ –≤ –æ—Ç—Ç–æ–∫).

–¢–∞–∫–∂–µ –≤—ã–ø–æ–ª–Ω–∏–º ¬´—Ü–µ–Ω—Ç—Ä–æ–≤–∫—É¬ª –∏ ¬´—Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–∞—Ü–∏—é¬ª –¥–∞–Ω–Ω—ã—Ö –ø—É—Ç—ë–º –∏—Ö –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏—è. –ú–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏–µ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è –∏–∑-–∑–∞ —Ç–æ–≥–æ, —á—Ç–æ —Ä–∞–∑–Ω—ã–µ –≤–µ–ª–∏—á–∏–Ω—ã –≤—ã—Ä–∞–∂–µ–Ω—ã –≤ —Ä–∞–∑–Ω—ã—Ö –µ–¥–∏–Ω–∏—Ü–∞—Ö –∏–∑–º–µ—Ä–µ–Ω–∏—è. –≠—Ç–æ –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã–π —ç—Ç–∞–ø –¥–ª—è —Ä–∞–±–æ—Ç—ã —Å –∞–ª–≥–æ—Ä–∏—Ç–º–æ–º K-means, –∏ —Å–æ–≥–ª–∞—Å–Ω–æ –∑–∞–ø–∏—Å—è–º –≤ —Å–µ—Ç–∏, –∞–ª–≥–æ—Ä–∏—Ç–º—ã –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –∏ —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞ —Ç–∞–∫–∂–µ –º–æ–≥—É—Ç —Ä–∞–±–æ—Ç–∞—Ç—å —Ç–æ—á–Ω–µ–µ –ø–æ—Å–ª–µ –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏—è –ø–æ–∫–∞–∑–∞—Ç–µ–ª–µ–π. –í —Ä–∞–º–∫–∞—Ö —É—á–µ–±–Ω–æ–≥–æ –ø—Ä–æ–µ–∫—Ç–∞ —Å—Ä–∞–≤–Ω–∏–º –º–µ—Ç—Ä–∏–∫–∏ –Ω–∞ —Å—ã—Ä—ã—Ö –∏ –Ω–∞ –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 

```python
# –º–∞—Å—à—Ç–∞–±–∏—Ä—É–µ–º –¥–∞–Ω–Ω—ã–µ
scaler = StandardScaler()
X_train_st = scaler.fit_transform(X_train)
X_test_st = scaler.transform(X_test)
```

## 3.2 –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏

–û–±—É—á–∏—Ç–µ –º–æ–¥–µ–ª—å –Ω–∞ train-–≤—ã–±–æ—Ä–∫–µ –¥–≤—É–º—è —Å–ø–æ—Å–æ–±–∞–º–∏:
–ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–µ–π,
—Å–ª—É—á–∞–π–Ω—ã–º –ª–µ—Å–æ–º.

–°–Ω–∞—á–∞–ª–∞ –æ–±—É—á–∏–º –º–æ–¥–µ–ª—å –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –Ω–∞ —Å—ã—Ä—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 

```python
# –∑–∞–¥–∞–µ–º –∞–ª–≥–æ—Ä–∏—Ç–º –¥–ª—è –º–æ–¥–µ–ª–∏ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
lr_model = LogisticRegression(random_state=0)

# –æ–±—É—á–∞–µ–º –º–æ–¥–µ–ª—å –Ω–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ
lr_model.fit(X_train, y_train)

# –ø—Ä–æ–≤–µ—Ä—è–µ–º –º–æ–¥–µ–ª—å –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ
lr_predictions = lr_model.predict(X_test)

# –≤–æ—Å–ø–æ–ª—å–∑—É–µ–º—Å—è —É–∂–µ –æ–±—É—á–µ–Ω–Ω–æ–π –º–æ–¥–µ–ª—å—é, —á—Ç–æ–±—ã —Å–¥–µ–ª–∞—Ç—å –ø—Ä–æ–≥–Ω–æ–∑—ã
lr_probabilities = lr_model.predict_proba(X_test)[:,1]
```

–ó–∞—Ç–µ–º –æ–±—É—á–∏–º –Ω–∞ –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö: 

```python
# –∑–∞–¥–∞–µ–º –∞–ª–≥–æ—Ä–∏—Ç–º –¥–ª—è –º–æ–¥–µ–ª–∏ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
lr_model_st = LogisticRegression(random_state=0)

# –æ–±—É—á–∞–µ–º –º–æ–¥–µ–ª—å –Ω–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ
lr_model_st.fit(X_train_st, y_train)

# –ø—Ä–æ–≤–µ—Ä—è–µ–º –º–æ–¥–µ–ª—å –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ
lr_predictions_st = lr_model_st.predict(X_test_st)

# –≤–æ—Å–ø–æ–ª—å–∑—É–µ–º—Å—è —É–∂–µ –æ–±—É—á–µ–Ω–Ω–æ–π –º–æ–¥–µ–ª—å—é, —á—Ç–æ–±—ã —Å–¥–µ–ª–∞—Ç—å –ø—Ä–æ–≥–Ω–æ–∑—ã
lr_probabilities_st = lr_model_st.predict_proba(X_test_st)[:,1]
```

–¢–µ–ø–µ—Ä—å –æ–±—É—á–∏–º –º–æ–¥–µ–ª—å –Ω–∞ –∞–ª–≥–æ—Ä–∏—Ç–º–µ —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞:

```python
# –æ–±—É—á–∞–µ–º –Ω–∞ —Å—ã—Ä—ã—Ö –¥–∞–Ω–Ω—ã—Ö
rf_model = RandomForestClassifier(n_estimators = 100, random_state = 0) 
rf_model.fit(X_train, y_train)
rf_predictions = rf_model.predict(X_test)
rf_probabilities = rf_model.predict_proba(X_test)[:,1]

# –æ–±—É—á–∞–µ–º –Ω–∞ –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö

rf_model_st = RandomForestClassifier(n_estimators = 100, random_state = 0) 
rf_model_st.fit(X_train_st, y_train)
rf_predictions_st = rf_model_st.predict(X_test_st)
rf_probabilities_st = rf_model_st.predict_proba(X_test_st)[:,1]
```

## 3.3 –û—Ü–µ–Ω–∫–∞ –º–µ—Ç—Ä–∏–∫ –º–æ–¥–µ–ª–∏

–û—Ü–µ–Ω–∏—Ç–µ –º–µ—Ç—Ä–∏–∫–∏ accuracy, precision –∏ recall –¥–ª—è –æ–±–µ–∏—Ö –º–æ–¥–µ–ª–µ–π –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ. –°—Ä–∞–≤–Ω–∏—Ç–µ –ø–æ –Ω–∏–º –º–æ–¥–µ–ª–∏. –ö–∞–∫–∞—è –º–æ–¥–µ–ª—å –ø–æ–∫–∞–∑–∞–ª–∞ —Å–µ–±—è –ª—É—á—à–µ –Ω–∞ –æ—Å–Ω–æ–≤–∞–Ω–∏–∏ –º–µ—Ç—Ä–∏–∫?

–ú–µ—Ç—Ä–∏–∫–∞ **accuracy**:    
* –¥–æ–ª—è –≤–µ—Ä–Ω–æ —É–≥–∞–¥–∞–Ω–Ω—ã—Ö –æ—Ç–≤–µ—Ç–æ–≤ –∏–∑ –≤—Å–µ—Ö –ø—Ä–æ–≥–Ω–æ–∑–æ–≤. –ß–µ–º –±–ª–∏–∂–µ –∑–Ω–∞—á–µ–Ω–∏–µ accuracy –∫ 100%, —Ç–µ–º –ª—É—á—à–µ.  
* –ú–µ—Ç—Ä–∏–∫—É —Ä–∞—Å—Å—á–∏—Ç—ã–≤–∞—é—Ç —Ñ—É–Ω–∫—Ü–∏–µ–π accuracy_score –∏–∑ –º–æ–¥—É–ª—è metrics. –ù–∞ –≤—Ö–æ–¥ —Ñ—É–Ω–∫—Ü–∏—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –≤–µ—Ä–Ω—ã–µ –∏ —Å–ø—Ä–æ–≥–Ω–æ–∑–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –∫–ª–∞—Å—Å–æ–≤ –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ.  
* accuracy —Ä–∞–±–æ—Ç–∞–µ—Ç –Ω–µ –≤—Å–µ–≥–¥–∞, –∞ —Ç–æ–ª—å–∫–æ –ø—Ä–∏ —É—Å–ª–æ–≤–∏–∏ –±–∞–ª–∞–Ω—Å–∞ –∫–ª–∞—Å—Å–æ–≤ ‚Äî –∫–æ–≥–¥–∞ –æ–±—ä–µ–∫—Ç–æ–≤ –∫–∞–∂–¥–æ–≥–æ –∫–ª–∞—Å—Å–∞ –ø—Ä–∏–º–µ—Ä–Ω–æ –ø–æ—Ä–æ–≤–Ω—É, 50% : 50%.

–¢.–∫. –≤ –Ω–∞—à–µ–º –¥–∞—Ç–∞—Å–µ—Ç–µ –¥–æ–ª—è –æ—Ç—Ç–æ–∫–∞ 27%, —Ç–æ –±–∞–ª–∞–Ω—Å –∫–ª–∞—Å—Å–æ–≤ –Ω–µ —Å–æ–±–ª—é–¥–∞–µ—Ç—Å—è - –ø—Ä–∏ –æ—Ü–µ–Ω–∫–µ –º–æ–¥–µ–ª–∏ –±–∏–Ω–∞—Ä–Ω–æ–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –æ—Ä–∏–µ–Ω—Ç–∏—Ä–æ–≤–∞—Ç—å—Å—è –Ω–∞ –¥—Ä—É–≥–∏–µ –º–µ—Ç—Ä–∏–∫–∏. 

–ú–µ—Ç—Ä–∏–∫–∞ **precision** (—Ç–æ—á–Ω–æ—Å—Ç—å):
* –¥–æ–ª—è –ø—Ä–∞–≤–∏–ª—å–Ω—ã—Ö –æ—Ç–≤–µ—Ç–æ–≤ —Ç–æ–ª—å–∫–æ —Å—Ä–µ–¥–∏ —Ü–µ–ª–µ–≤–æ–≥–æ –∫–ª–∞—Å—Å–∞ (TP / TP + FP)
* –ø—Ä–∏–º–µ—Ä: –≤—ã –Ω–µ —Ö–æ—Ç–∏—Ç–µ, —á—Ç–æ–±—ã –º–æ–¥–µ–ª—å —á–∞—Å—Ç–æ ¬´—Å—Ä–∞–±–∞—Ç—ã–≤–∞–ª–∞ –ø–æ–ø—É—Å—Ç—É¬ª. –ù–∞–ø—Ä–∏–º–µ—Ä, –∫–æ–≥–¥–∞ –≤—ã –ø—Ä–∏–Ω–∏–º–∞–µ—Ç–µ —Ä–µ—à–µ–Ω–∏–µ, –≤—ã–¥–∞–≤–∞—Ç—å –∫–ª–∏–µ–Ω—Ç—É –∫—Ä–µ–¥–∏—Ç –∏–ª–∏ –Ω–µ—Ç, —Å–∏—Ç—É–∞—Ü–∏—è –Ω–µ–≤–æ–∑–≤—Ä–∞—Ç–∞ –±–æ–ª–µ–µ –Ω–µ–ø—Ä–∏—è—Ç–Ω–∞, —á–µ–º –Ω–µ–≤—ã–¥–∞—á–∞.


–ú–µ—Ç—Ä–∏–∫–∞ **recall**:
* –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç, —Å–∫–æ–ª—å–∫–æ —Ä–µ–∞–ª—å–Ω—ã—Ö –æ–±—ä–µ–∫—Ç–æ–≤ "1" –∫–ª–∞—Å—Å–∞ –≤—ã —Å–º–æ–≥–ª–∏ –æ–±–Ω–∞—Ä—É–∂–∏—Ç—å —Å –ø–æ–º–æ—â—å—é –º–æ–¥–µ–ª–∏. 
* –º–µ—Ç—Ä–∏–∫–∞ –ø–æ–ª–µ–∑–Ω–∞ –ø—Ä–∏ –¥–∏–∞–≥–Ω–æ—Å—Ç–∏–∫–µ –∑–∞–±–æ–ª–µ–≤–∞–Ω–∏–π: –ª—É—á—à–µ –æ—Ç–ø—Ä–∞–≤–∏—Ç—å –ø–∞—Ü–∏–µ–Ω—Ç–∞ –Ω–∞ –ø–æ–≤—Ç–æ—Ä–Ω–æ–µ –æ–±—Å–ª–µ–¥–æ–≤–∞–Ω–∏–µ –∏ —É–∑–Ω–∞—Ç—å, —á—Ç–æ —Ç—Ä–µ–≤–æ–≥–∞ –±—ã–ª–∞ –ª–æ–∂–Ω–æ–π, —á–µ–º –ø—Ä–æ–∑–µ–≤–∞—Ç—å –Ω–∞—Å—Ç–æ—è—â–∏–π –¥–∏–∞–≥–Ω–æ–∑.

–ú–µ—Ç—Ä–∏–∫–∞ **f1-score**:
* –°–≤–æ–¥–Ω–∞—è –º–µ—Ç—Ä–∏–∫–∞ (2 * precision * recall / (precision + recall))
* –§—É–Ω–∫—Ü–∏—è —Ç–∞–∫–∂–µ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –æ–¥–Ω–æ —á–∏—Å–ª–æ –æ—Ç 0 –¥–æ 1. –ß–µ–º –±–ª–∏–∂–µ –∫ –µ–¥–∏–Ω–∏—Ü–µ, —Ç–µ–º –ª—É—á—à–µ.

```python
# —Ñ—É–Ω–∫—Ü–∏—è, –∫–æ—Ç–æ—Ä–∞—è –≤—ã–≤–æ–¥–∏—Ç –º–µ—Ç—Ä–∏–∫–∏ –¥–ª—è –º–æ–¥–µ–ª–µ–π –±–∏–Ω–∞—Ä–Ω–æ–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏, 
# –Ω–∞ –≤—Ö–æ–¥–µ –ø–æ–ª—É—á–∞–µ—Ç —Ü–µ–ª–µ–≤—É—é –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏ –∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –º–æ–¥–µ–ª–∏
def show_metrics(y_test, predictions):
    # –≤—ã–≤–æ–¥–∏–º –Ω–∞ —ç–∫—Ä–∞–Ω –∏–Ω—Ç–µ—Ä–µ—Å—É—é—â–∏–µ –º–µ—Ç—Ä–∏–∫–∏
    print('Accuracy: {:.2f}'.format(accuracy_score(y_test, predictions)))
    print('Precision: {:.2f}'.format(precision_score(y_test, predictions)))
    print('Recall: {:.2f}'.format(recall_score(y_test, predictions)))
    print('F1: {:.2f}'.format(f1_score(y_test, predictions)))
            
    return
```

–û—Ü–µ–Ω–∏–º –º–µ—Ç—Ä–∏–∫–∏ –¥–ª—è –º–æ–¥–µ–ª–µ–π:

```python
for pred in [lr_predictions, lr_predictions_st, rf_predictions, rf_predictions_st]:
    show_metrics(y_test, pred)
    print()
```

Accuracy:
* –ü—Ä–∏ –¥–∏—Å–±–∞–ª–∞–Ω—Å–µ –∫–ª–∞—Å—Å–æ–≤ –º–µ—Ç—Ä–∏–∫–µ —É–¥–µ–ª—è–µ–º –Ω–µ–º–Ω–æ–≥–æ –º–µ–Ω—å—à–µ –≤–Ω–∏–º–∞–Ω–∏—è
* –•—É–∂–µ –ø–æ–∫–∞–∑–∞–ª–∞ —Å–µ–±—è –º–æ–¥–µ–ª—å –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –Ω–∞ —Å—ã—Ä—ã—Ö –¥–∞–Ω–Ω—ã—Ö.

Precision:
* –õ—É—á—à–µ –≤—Å–µ–≥–æ –ø–æ–∫–∞–∑–∞–ª–∞ —Å–µ–±—è –º–æ–¥–µ–ª—å –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –Ω–∞ –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö.
* –°–ª—É—á–∞–π–Ω—ã–π –ª–µ—Å –Ω–∞ –ª—é–±—ã—Ö –¥–∞–Ω–Ω—ã—Ö –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç –∑–Ω–∞—á–µ–Ω–∏–µ 0.84. –í–æ–∑–º–æ–∂–Ω–æ, –ø—Ä–∏ –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã—Ö —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞—Ö —Å —á–∏—Å–ª–æ–º –¥–µ—Ä–µ–≤—å–µ–≤ –∏–ª–∏ –≥–ª—É–±–∏–Ω–æ–π, –º–æ–∂–Ω–æ –±—ã–ª–æ –±—ã –¥–æ–±–∏—Ç—å—Å—è –±–æ–ª–µ–µ –≤—ã—Å–æ–∫–∏—Ö –ø–æ–∫–∞–∑–∞—Ç–µ–ª–µ–π. 

Recall:
* –õ—É—á—à–µ –≤—Å–µ–≥–æ –ø–æ–∫–∞–∑–∞–ª–∞ —Å–µ–±—è –º–æ–¥–µ–ª—å –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –Ω–∞ –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö.
* –•—É–∂–µ –ø–æ–∫–∞–∑–∞–ª–∞ —Å–µ–±—è –º–æ–¥–µ–ª—å –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –Ω–∞ —Å—ã—Ä—ã—Ö –¥–∞–Ω–Ω—ã—Ö.

–û—Ü–µ–Ω–∏–º, –∫–∞–∫–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ –∫–∞–∂–¥–∞—è –º–æ–¥–µ–ª—å —Å—á–∏—Ç–∞–µ—Ç —Å–∞–º—ã–º–∏ –≤–∞–∂–Ω—ã–º–∏: 

```python
# –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –∞—Ç—Ä–∏–±—É—Ç coef_ –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º, –æ—Ç—Å–æ—Ä—Ç–∏—Ä—É–µ–º –ø–æ —É–±—ã–≤–∞–Ω–∏—é, –≤—ã–≤–µ–¥–µ–º –Ω–∞ —ç–∫—Ä–∞–Ω
features_lr = pd.DataFrame(lr_model_st.coef_.T, X.columns).reset_index()
features_lr.columns = ['feature', 'coef']
features_lr['coef'] = features_lr['coef'].apply(lambda x: abs(x))
features_lr = features_lr.sort_values(by='coef', ascending=False)
print('–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –≤ –º–æ–¥–µ–ª–∏ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏:')
print(features_lr)
```

## 4.1 –°—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–∞—Ü–∏—è –¥–∞–Ω–Ω—ã—Ö

–ü–æ–¥–≥–æ—Ç–æ–≤–∏–º –¥–∞–Ω–Ω—ã–µ –¥–ª—è —Ä–∞–±–æ—Ç—ã —Å –∞–ª–≥–æ—Ä–∏—Ç–º–æ–º K-Means. –°–Ω–∞—á–∞–ª–∞ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –ø—Ä–æ–≤–µ—Å—Ç–∏ –æ–±—è–∑–∞—Ç–µ–ª—å–Ω—É—é —Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–∞—Ü–∏—é –¥–∞–Ω–Ω—ã—Ö –Ω–∞ –º–∞—Ç—Ä–∏—Ü–µ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ (churn —É–∂–µ –æ—Ç–±—Ä–æ—à–µ–Ω):

```python
X_st = scaler.fit_transform(X)
```

## 4.2 –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –º–∞—Ç—Ä–∏—Ü—ã —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–π –∏ –¥–µ–Ω–¥—Ä–æ–≥—Ä–∞–º–º—ã.

```python
# –í –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π linked —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∞ —Ç–∞–±–ª–∏—Ü–∞ ¬´—Å–≤—è–∑–æ–∫¬ª –º–µ–∂–¥—É –æ–±—ä–µ–∫—Ç–∞–º–∏. –ï—ë –º–æ–∂–Ω–æ –≤–∏–∑—É–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å –∫–∞–∫ –¥–µ–Ω–¥—Ä–æ–≥—Ä–∞–º–º—É
linked = linkage(X_st, method='ward')

# –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –¥–µ–Ω–¥—Ä–æ–≥—Ä–∞–º–º—ã 
plt.figure(figsize=(15, 10))  
dendrogram(linked, orientation='top')
plt.title('–ò–µ—Ä–∞—Ä—Ö–∏—á–µ—Å–∫–∞—è –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏—è')
plt.show() 
```

* –ù–∞ –¥–µ–Ω–¥—Ä–æ–≥—Ä–∞–º–º–µ —è–≤–Ω–æ –≤—ã–¥–µ–ª–µ–Ω—ã 4 –∫–ª–∞—Å—Ç–µ—Ä–∞ —Ä–∞–∑–Ω–æ–≥–æ —Ü–≤–µ—Ç–∞. 
* 4-—ã–π –∫–ª–∞—Å—Ç–µ—Ä —Å–∞–º—ã–π –±–æ–ª—å—à–æ–π, –≤–æ–∑–º–æ–∂–Ω–æ, –æ–Ω —Ä–∞–∑–±–∏–≤–∞–µ—Ç—Å—è –Ω–∞ –±–æ–ª—å—à–µ–µ —á–∏—Å–ª–æ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤. 
* –ü–æ —É—Å–ª–æ–≤–∏—è–º, –º—ã –¥–æ–≥–æ–≤–æ—Ä–∏–º—Å—è –∑–∞ —á–∏—Å–ª–æ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤ –ø—Ä–∏–Ω—è—Ç—å n=5, —á—Ç–æ–±—ã —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –º–æ–∂–Ω–æ –±—ã–ª–æ —Å—Ä–∞–≤–Ω–∏–≤–∞—Ç—å —Å —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º–∏ –æ—Å—Ç–∞–ª—å–Ω—ã—Ö —Å—Ç—É–¥–µ–Ω—Ç–æ–≤.

## 4.3 –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏ –∫–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏ –Ω–∞ –∞–ª–≥–æ—Ä–∏—Ç–º–µ K-Means

```python
# –∑–∞–¥–∞—ë–º —á–∏—Å–ª–æ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤, —Ä–∞–≤–Ω–æ–µ 5, –∏ —Ñ–∏–∫—Å–∏—Ä—É–µ–º –∑–Ω–∞—á–µ–Ω–∏–µ random_state –¥–ª—è –≤–æ—Å–ø—Ä–æ–∏–∑–≤–æ–¥–∏–º–æ—Å—Ç–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞
km = KMeans(n_clusters=5, random_state=0) 

# –ø—Ä–∏–º–µ–Ω—è–µ–º –∞–ª–≥–æ—Ä–∏—Ç–º –∫ –¥–∞–Ω–Ω—ã–º –∏ —Ñ–æ—Ä–º–∏—Ä—É–µ–º –≤–µ–∫—Ç–æ—Ä –∫–ª–∞—Å—Ç–µ—Ä–æ–≤ 
# –ø—Ä–æ–≥–Ω–æ–∑–∏—Ä—É–µ–º –∫–ª–∞—Å—Ç–µ—Ä—ã –¥–ª—è –Ω–∞–±–ª—é–¥–µ–Ω–∏–π (–∞–ª–≥–æ—Ä–∏—Ç–º –ø—Ä–∏—Å–≤–∞–∏–≤–∞–µ—Ç –∏–º –Ω–æ–º–µ—Ä–∞ –æ—Ç 0 –¥–æ 4)
labels = km.fit_predict(X_st)

# —Å–æ—Ö—Ä–∞–Ω—è–µ–º –º–µ—Ç–∫–∏ –∫–ª–∞—Å—Ç–µ—Ä–∞ –≤ –ø–æ–ª–µ –Ω–∞—à–µ–≥–æ –¥–∞—Ç–∞—Å–µ—Ç–∞
df['cluster_km'] = labels
```
